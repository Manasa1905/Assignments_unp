6.Check a number is prime or not

```python
s=0
n=int(input("enter a number"))
for i in range(1,n-1):
  if(n%i)==0:
    s=s+1
if(s==1):
  print("it is a prime number")
else:
  print("it is not a prime number")
```

8.Print the series of first n fibonacci numbers
